```python
import math

def eval_fun(x):
    return x * (x - 1.5)
print(eval_fun(0.7),eval_fun(0.8))
def exhaustive_search_method(eval_fun, a, b, acc):
    n= math.ceil((100/acc) -1) + 1
    h = (b - a) / n
    EPS = 0
    print(h)
    for i in range(n):
        prev_point = a + (i+1) * h
        next_point = a + (i + 2) * h
        f_prev = eval_fun(prev_point)
        f_next = eval_fun(next_point)
        
        print(f"Checking [{prev_point:.4f}, {next_point:.4f}]: f_prev={f_prev:.4f}, f_next={f_next:.4f}")
        
        if abs(f_next - f_prev) <= EPS or f_next > f_prev:
            mid = (prev_point + next_point) / 2
            print(f"Final IU: [{prev_point:.4f}, {next_point:.4f}], midpoint={mid:.4f}")
            return (prev_point, next_point, mid)
    
    print("No minimum detected; function may be monotonic.")
    return None
result = exhaustive_search_method(eval_fun, 0, 1, 10)  
print(result)
```

```python
def eval_fun(x):
    return x * (x - 1.5) * (x - 2.5)
result = exhaustive_search_method(eval_fun, 1.9, 2.2, 1)   
print(result)
```

```python
def eval_fun(x):
    return (x**4) - 3*(x**3) + 2*x + 1
result = exhaustive_search_method(eval_fun, 2, 3, 0.1)   
print(result)
```
